```python
#替换为实际按钮xpath
chaxun=r'/html/body/div[1]/div/div[3]/div[1]/div/div[3]/div/div/div/section/div/div/div/div[1]/form/div/div[15]/button[1]'
xuanke=r'/html/body/div[1]/div/div[3]/div[1]/div/div[3]/div/div/div/section/div/div/div/div[2]/div/div[1]/div[3]/table/tbody/tr/td[1]/div/button'
#替换为你需要抢的课程的教师
teacher_name='柳文章'

select_lesson2=driver.find_element(By.XPATH,chaxun)
select_lesson2.click()
#选课按钮的xpath
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, xuanke)))
choose=driver.find_element(By.XPATH,xuanke)

courses = []
def update_course_rows():
    html_content=driver.page_source
    html_content=BeautifulSoup(html_content,'html.parser')
    course_rows = html_content.find_all('tr', class_='el-table__row')
    
    for row in course_rows:
        course = {}
        cells = row.find_all('td', class_='el-table__cell')
        
        if len(cells) > 6:
            # xpath = "//button[contains(@class, 'el-button course-select')]"
            # html_tree = etree.HTML(str(cells[0]))
            # button_elements = html_tree.xpath(xpath)
            # button_tree = etree.ElementTree(button_elements[0])

            # 获取按钮元素的全局 XPath
            # global_xpath = button_tree.getpath(button_elements[0])
            # button = global_xpath
            # button_xpath = etree.ElementTree(etree.HTML(str(button))).getpath(button)

            course_name = cells[2].find('div', class_='text-color-3').text.strip()
            course_code = cells[3].find('div', class_='text-color-9').text.strip()
            teacher_name = cells[4].find('div', class_='text-color-3').text.strip()
            location = cells[5].find('div', class_='text-color-9').text.strip()
            people_div = cells[6].find('div').text.strip()
            

            # course['button_xpath'] = button
            course['name'] = course_name
            course['code'] = course_code
            course['teacher'] = teacher_name
            course['location'] = location
            course['people'] = people_div
            courses.append(course)
# 打印课程信息
plan_time= datetime(2024, 12, 11, 13, 00, 00)#设置抢课起始时间
while(1):
    now = datetime.now()
    if now >= plan_time:
         break
    time.sleep(1)
    print('时间未到',now)
update_course_rows()
wait=1
cnt=0
while True:
    try:
        while True: 
            print('刷新中')
            print("-" * 30)
            time.sleep(2)
            for course in courses:
                    if teacher_name in course['teacher']:
                            if number_compare(course['people']):
                                choose.click()
                                # print(f"按钮XPath: {course['button_xpath']}")
                                print(f"课程名称: {course['name']}")
                                print(f"课程代码: {course['code']}")
                                print(f"教师: {course['teacher']}")
                                print(f"地点: {courses['location']}")
                                print(f"人数: {course['people']}")
                                print("-" * 30)
                                # pyautogui.click(x, y)
                                # pyautogui.click(x, y)
                                
                                print('选课成功')
                                wait=0
                                break
                                
                            else:
                                select_lesson2.click()#chauxn按钮
                                time.sleep(2)
                                print(f"人数: {course['people']}")
                                print('出错次数',cnt)
                                now = datetime.now()
                                print(now)
                                courses=[]
                                update_course_rows()
                                
                                break
                                # print(f"课程名称: {course['name']}")
                                # print(f"人数: {course['people']}")
    
            if wait==0:
                exit()
                break
    except Exception as e:
                try:    
                    print('出错了')   #当提示操作频繁时
                    time.sleep(2)
                    re=driver.find_element(By.XPATH,'/html/body/div[10]/div/div[3]/button')
                    time.sleep(1)
                    re.click()
                    
                    cnt+=1
                    print('重新刷新',cnt)
                except Exception as e:
                    print('网络连接异常',e) #当网络连接异常时
                    time.sleep(2)
```
